```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,6]

plt.plot(x,y,color='g')
#add title and label names,,,,ll
plt.title("test plot")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Multiple lines
x = [1,2,3]
y = [1,4,6]
z = [4,3,2]
plt.plot(x,y)
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y and z')
#legend function takes a list as the argument
plt.legend(['y','z'])
plt.show()
```

```python
CountryCount=df[df.Year==2018].groupby('Region').Country.count()
CountryCount
```

```python
plt.pie(CountryCount,labels=CountryCount.index,autopct='%1.1f%%')
plt.axis('equal')
plt.rcParams["figure.figsize"]=10,6
plt.show()
```

```python
#Excercise 1:
Pop_by_Region=df[df.Year==2018].groupby('Region').Population.sum()/10**3
Pop_by_Region
```

```python
plt.pie(Pop_by_Region,labels=Pop_by_Region.index,autopct='%1.1f%%')
plt.axis('equal')
plt.rcParams["figure.figsize"]=10,6
plt.show()

```

```python
GDP_By_Region=(df[df.Year==2018].groupby('Region')['GDPEKS'].sum()/10**3).sort_values()
GDP_By_Region
```

```python
import numpy as np
height= GDP_By_Region
bars= GDP_By_Region.index
y_pos=np.arange(len(bars))
plt.barh(y_pos, height,color='g')
plt.yticks(y_pos, bars)
plt.rcParams['axes.ymargin'] = 0
plt.show()
```

```python
GDP_By_Region.plot(kind='bar')
```

```python
plt.bar(GDP_By_Region.index,GDP_By_Region)
plt.xticks(rotation=90)
plt.rcParams['axes.xmargin'] = 0
plt.show()
```

```python
US=df[df.Country=='United States']
China=df[df.Country=='China']
plt.plot(US.Year,US.Population/10**3)
plt.plot(China.Year,China.Population/10**3)
plt.legend(['US','China'])
plt.show()
```

```python
#Excercise 2: PopulationGrowth for Germany, China, US
Germany=df[df.Country=='Germany']
China=df[df.Country=='China']
US=df[df.Country=='United States']
plt.plot(Germany.Year,Germany.PopulationGrowth)
plt.plot(China.Year,China.PopulationGrowth)
plt.plot(US.Year,US.PopulationGrowth)
plt.legend(['Germany','China','US'])
plt.show()
```

```python
GDP2018=df[df.Year==2018].PerCapitaIncome
plt.hist(GDP2018,bins=20,color='salmon',edgecolor='black')
#plt.rcParams['axes.xmargin'] = 0
plt.show()
#plt.savefig('bars.png')
```

```python
Rich=df[(df.Year==2018) & (df.PerCapitaIncome>60000)]

Rich[['Country','Region','PerCapitaIncome']].sort_values(by='PerCapitaIncome',ascending=False)
```

```python
df_2018 = df[df.Year == 2018]
asia_2018 = df_2018[df_2018.Region == 'Asia']
europe_2018 = df_2018[df_2018.Region == 'Western Europe']


plt.subplot(2, 1, 1) # subplot(211)
plt.title('Distribution of GDP Per Capita')
plt.hist(asia_2018['PerCapitaIncome'], 20, range=(0, 50000), edgecolor='black')
plt.ylabel('Asia')
plt.subplot(2, 1, 2)
plt.hist(europe_2018['PerCapitaIncome'], 20, range=(0, 50000), edgecolor='black')
plt.ylabel('Europe')
plt.show()
```
